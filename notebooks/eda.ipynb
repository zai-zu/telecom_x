📂 # Extract data

```python
import pandas as pd
import os
import numpy as np
```

```python
data = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'

df_custumers = pd.read_json(data)

df_custumers
```

 🛠️ #Transformation

```python
df_custumers.columns
```

```python
df_custumers.info ()
```

```python
# Normalization: flatten nested columns in df_custumers
df_custumers_norm = pd.concat([ #contac 
	df_custumers.drop(['customer', 'phone', 'internet', 'account'], axis=1),
	pd.json_normalize(df_custumers['customer']),
	pd.json_normalize(df_custumers['phone']),
	pd.json_normalize(df_custumers['internet']),
	pd.json_normalize(df_custumers['account'])
], axis=1)
```

```python
df_custumers_norm
```

```python
#save raw data
output_dir_raw = '../data/raw'
file_path =os.path.join(output_dir_raw, 'custumers_norm.csv')
os.makedirs(output_dir_raw, exist_ok=True) 
df_custumers_norm.to_csv(file_path, index=False)
```

```python
# Check for consistency in categorical variables (e.g., 'gender', 'SeniorCitizen', 'Partner', etc.)
# Some variables should have only a limited set of values. Let's check if they match expectations
print("\nChecking Consistency in Categorical Variables:")
expected_values = {
    'gender': ['Male', 'Female'],
    'SeniorCitizen': [0, 1],  # Assuming SeniorCitizen is binary (0 or 1)
    'Partner': ['Yes', 'No'],
    'Dependents': ['Yes', 'No'],
    'PhoneService': ['Yes', 'No'],
    'MultipleLines': ['Yes', 'No', 'No phone service'],
    'InternetService': ['DSL', 'Fiber optic', 'No'],
    'OnlineSecurity': ['Yes', 'No', 'No internet service'],
    'OnlineBackup': ['Yes', 'No', 'No internet service'],
    'DeviceProtection': ['Yes', 'No', 'No internet service'],
    'TechSupport': ['Yes', 'No', 'No internet service'],
    'StreamingTV': ['Yes', 'No', 'No internet service'],
    'StreamingMovies': ['Yes', 'No', 'No internet service'],
    'Contract': ['Month-to-month', 'One year', 'Two year'],
    'PaperlessBilling': ['Yes', 'No'],
    'PaymentMethod': ['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'],
}
```

```python
# Check for any unexpected values in the categorical columns
for col, expected in expected_values.items():
    unique_vals = df_custumers_norm[col].unique()
    invalid_vals = set(unique_vals) - set(expected)
    if invalid_vals:
        print(f"Invalid values in {col}: {invalid_vals}")
    else:
        print(f"All values in {col} are valid.")
```

```python
# Handling Missing Values
# If there are any missing values, we can handle them by either filling them with the mean/median (for numerical columns)
# or the most frequent value (for categorical columns). Alternatively, you can drop rows with missing data if needed.

print("\nHandling Missing Values:")

# Fill missing numerical columns with the median value (common approach for skewed data)
numerical_columns = df_custumers_norm.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_columns:
    median_value = df_custumers_norm[col].median()
    df_custumers_norm[col].fillna(median_value, inplace=True)

# Fill missing categorical columns with the most frequent value (mode)
categorical_columns = df_custumers_norm.select_dtypes(include=['object']).columns
for col in categorical_columns:
    mode_value = df_custumers_norm[col].mode()[0]
    df_custumers_norm[col].fillna(mode_value, inplace=True)

print("Missing values handled.")

# Handling Duplicates
# We will drop the duplicate rows to ensure we only have unique entries.
print("\nHandling Duplicates:")
```

```python
df_custumers_norm.drop_duplicates(inplace=True)

print("Duplicates removed.")
```

```python
# Fixing Inconsistencies in Categorical Columns
# Some categorical variables may have invalid values or unexpected entries.
# We will replace those values with the valid ones, or drop rows that contain invalid values.

print("\nFixing Inconsistencies in Categorical Columns:")

# Expected values for categorical columns, as per the earlier check
expected_values = {
    'gender': ['Male', 'Female'],
    'SeniorCitizen': [0, 1],
    'Partner': ['Yes', 'No'],
    'Dependents': ['Yes', 'No'],
    'PhoneService': ['Yes', 'No'],
    'MultipleLines': ['Yes', 'No', 'No phone service'],
    'InternetService': ['DSL', 'Fiber optic', 'No'],
    'OnlineSecurity': ['Yes', 'No', 'No internet service'],
    'OnlineBackup': ['Yes', 'No', 'No internet service'],
    'DeviceProtection': ['Yes', 'No', 'No internet service'],
    'TechSupport': ['Yes', 'No', 'No internet service'],
    'StreamingTV': ['Yes', 'No', 'No internet service'],
    'StreamingMovies': ['Yes', 'No', 'No internet service'],
    'Contract': ['Month-to-month', 'One year', 'Two year'],
    'PaperlessBilling': ['Yes', 'No'],
    'PaymentMethod': ['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'],
}

# Loop through the columns and fix any invalid values
for col, valid_values in expected_values.items():
    # Replace invalid values with the most frequent (mode) valid value
    df_custumers_norm[col] = df_custumers_norm[col].apply(lambda x: x if x in valid_values else df_custumers_norm[col].mode()[0])

print("Inconsistencies in categorical columns fixed.")

# Handling Data Types
# Ensure that each column has the correct data type.
# We will convert 'SeniorCitizen' and 'Churn' to categorical values for consistency if they are not already.
print("\nHandling Data Types:")
```

```python
# Convert 'SeniorCitizen' and 'Churn' to categorical
df_custumers_norm['SeniorCitizen'] = df_custumers_norm['SeniorCitizen'].astype('category')
df_custumers_norm['Churn'] = df_custumers_norm['Churn'].astype('category')


```

```python
# Handle invalid values before converting to float
print("\nCleaning 'Charges.Total' column:")

# Replace non-numeric values (e.g., empty strings, or other invalid entries) with NaN
df_custumers_norm['Charges.Total'] = pd.to_numeric(df_custumers_norm['Charges.Total'], errors='coerce')

# Optionally, you can fill NaN values in 'Charges.Total' with the median or another value
df_custumers_norm['Charges.Total'].fillna(df_custumers_norm['Charges.Total'].median(), inplace=True)

# Now safely convert to float
df_custumers_norm['Charges.Total'] = df_custumers_norm['Charges.Total'].astype(float)

print("'Charges.Total' column cleaned and converted to float.")

# Proceed with the rest of the process
# Convert 'Charges.Monthly' to float if it hasn't been handled yet
df_custumers_norm['Charges.Monthly'] = pd.to_numeric(df_custumers_norm['Charges.Monthly'], errors='coerce')
df_custumers_norm['Charges.Monthly'].fillna(df_custumers_norm['Charges.Monthly'].median(), inplace=True)
df_custumers_norm['Charges.Monthly'] = df_custumers_norm['Charges.Monthly'].astype(float)

print("Both 'Charges.Monthly' and 'Charges.Total' columns are now cleaned and converted to float.")
```

```python
# Final Check (Optional but recommended)
# After handling missing values, duplicates, and inconsistencies, it's always good to perform a final check.
print("\nFinal Check:")
print(df_custumers_norm.info())  # This will show the data types and non-null counts
print("\nUnique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df_custumers_norm[col].unique()}")
```
