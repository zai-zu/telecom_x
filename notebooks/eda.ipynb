```python
import pandas as pd
import os
import numpy as np
```

```python
data = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'

df_custumers = pd.read_json(data)

df_custumers
```

 üõ†Ô∏è #Transformation

```python
df_custumers.columns
```

```python
df_custumers.info ()
```

```python
# Normalization: flatten nested columns in df_custumers
df_custumers_norm = pd.concat([ #contac 
	df_custumers.drop(['customer', 'phone', 'internet', 'account'], axis=1),
	pd.json_normalize(df_custumers['customer']),
	pd.json_normalize(df_custumers['phone']),
	pd.json_normalize(df_custumers['internet']),
	pd.json_normalize(df_custumers['account'])
], axis=1)
```

```python
df_custumers_norm
```

```python
#save raw data
output_dir_raw = '../data/raw'
file_path =os.path.join(output_dir_raw, 'custumers_norm.csv')
os.makedirs(output_dir_raw, exist_ok=True) 
df_custumers_norm.to_csv(file_path, index=False)
```
